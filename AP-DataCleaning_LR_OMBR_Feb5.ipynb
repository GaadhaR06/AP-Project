```python
from statsmodels.stats.outliers_influence import variance_inflation_factor  
```

```python
def computeVIF(data):
    vif = pd.DataFrame()
    # vif is a blank dataframe
    vif['Feature'] = data.columns
    vif['Value'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    print(vif)

computeVIF(X_1)
```

```python
X_lr_1.to_excel('X.xlsx')
```

```python
X_lr_1.columns = X_lr_1.columns.str.replace('[^A-Za-z0-9_()]+', '_', regex=True)
X_lr_1.columns
```

```python
from sklearn.model_selection import train_test_split
X_lr_train,X_lr_test,y_lr_train,y_lr_test=train_test_split(X_lr_1,y_lr,test_size=0.2,random_state=10)
X_lr_train.shape,X_lr_test.shape,y_lr_train.shape,y_lr_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logR= LogisticRegression(random_state=10)
logR.fit(X_lr_train,y_lr_train)
```

```python
y_lr_pred=logR.predict(X_lr_test)
y_lr_pred
```

```python
y_lr_pred_prob=logR.predict_proba(X_lr_test)
y_lr_pred_prob
```

```python
from sklearn.metrics import confusion_matrix,classification_report, roc_curve,roc_auc_score

conf_matrix= confusion_matrix(y_lr_test,y_lr_pred)
print('The confusion matrix:\n',conf_matrix)

class_report=classification_report(y_lr_test,y_lr_pred)
print('The classification report:\n',class_report)
```

```python
fpr,tpr,threshold=roc_curve(y_lr_test,y_lr_pred_prob[:,1])
plt.plot(fpr,tpr);
```

```python
score=roc_auc_score(y_lr_test,y_lr_pred)
print('The ROC-AUC-Score:',score)
```

```python
from sklearn.model_selection import cross_val_score
# Perform 5-fold cross-validation
scores = cross_val_score(logR, X_lr_train, y_lr_train, cv=5, scoring='accuracy')

# Print the cross-validation scores and mean accuracy
print("Cross-validation scores:", scores)
print("Mean Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_predict

# Get cross-validated predictions
y_pred = cross_val_predict(logR, X_lr_train, y_lr_train, cv=5)

# Now you can evaluate using metrics
from sklearn.metrics import accuracy_score
print("Cross-validated Accuracy:", accuracy_score(y_lr_train, y_pred))
```

# OMBD

```python
terr_ombd=terr[terr['doubtterr']==1]
terr_ombd.shape
```

```python
df= terr_ombd[terr_ombd['provstate']== 'Andhra pradesh']
df
```

```python
terr_ombd['provstate'] = terr_ombd['provstate'].replace('Andhra pradesh', 'Andhra Pradesh')
terr_ombd['provstate'] = terr_ombd['provstate'].replace('Orissa','Odisha')
terr_ombd=terr_ombd[terr_ombd['provstate'] != 'Unknown']
```

```python
terr_ombd=terr_ombd[['provstate','city','latitude','longitude','nkill','nwound']]
terr_ombd.shape
```

```python
terr_ombd['casualities']=terr_ombd['nkill']+terr_ombd['nwound']
terr_ombd.fillna(0,inplace=True)
terr_ombd.head()
```

```python
terr_ombd.drop(['nkill','nwound'],axis=1,inplace=True)
terr_ombd
```

```python
result = terr_ombd.groupby(["provstate", "city", "latitude", "longitude"], as_index=False)["casualities"].sum()

# Rename columns for clarity
#result.rename(columns={"provstate": "State", "city": "City", "latitude": "Latitude", "longitude": "Longitude", "casualities": "casualities"}, inplace=True)
result
```

```python
terr_ombd.columns
```

```python
result = terr_ombd.groupby("provstate").apply(lambda x: pd.Series({
    "OutputLatitude": (x["latitude"] * x["casualities"]).sum() / x["casualities"].sum(),
    "OutputLongitude": (x["longitude"] * x["casualities"]).sum() / x["casualities"].sum(),
      "casualities": x["casualities"].sum()
})).reset_index()

print(result)
```

```python
result.to_excel('terr_ombd_output.xlsx')
```
