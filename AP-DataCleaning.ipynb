```python
terr = terr.dropna(how = "any")
```

```python
terr = terr.drop(['Unnamed: 0'], axis = 1)
```

```python
terr = terr[~terr.eq("Unknown").any(axis=1)]
```

```python
terr = terr.drop(['nkill', 'n-kill-cat'], axis = 1)
```

```python
terr
```

```python
y = terr['n-kill-cat-num']
y
```

```python
X = terr.drop(['n-kill-cat-num'], axis = 1)
X
```

```python
X_1=pd.get_dummies(X,dtype=int)
X_1
```

```python
X_1=sm.add_constant(X_1)
X_1
```

```python
from sklearn.model_selection import train_test_split
X_train_1,X_test_1, y_train_1, y_test_1=train_test_split(X_1,y,test_size=0.2,random_state=10)
X_train_1.shape,X_test_1.shape, y_train_1.shape, y_test_1.shape
```

```python
mlr_1=sm.OLS(y_train_1,X_train_1)
```

```python
mlr_1=mlr_1.fit()
```

```python
mlr_1.params
```

```python
mlr_1.summary()
```

```python
correlation_matrix = X_1.corr()
correlation_matrix
```

```python
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Find features with correlation > |0.6|
high_correlation = [column for column in upper.columns if any(upper[column] > 0.6)]

print("Highly Correlated Features to Drop:", high_correlation)
```

```python
X_2 = X_1.drop(columns=high_correlation)
```

```python
X_2
```

```python
X_train_2,X_test_2, y_train_2, y_test_2=train_test_split(X_2,y,test_size=0.2,random_state=10)
X_train_2.shape,X_test_2.shape, y_train_2.shape, y_test_2.shape
```

```python
mlr_2 = sm.OLS(y_train_2, X_train_2)
mlr_2 = mlr_2.fit()
```

```python
mlr_2.summary()
```

```python
# computeVIF(X_2)
```

```python
correlation_matrix2 = X_2.corr()
correlation_matrix2
```

```python
upper2 = correlation_matrix2.where(np.triu(np.ones(correlation_matrix2.shape), k=1).astype(bool))

# Find features with correlation > |0.6|
high_correlation2 = [column for column in upper2.columns if any(upper2[column] > 0.6)]

print("Highly Correlated Features to Drop:", high_correlation2)
```

```python
# Fitting Multiple Linear Regression model 
mlr_2_summary = mlr_2.summary()

# Extract p-values directly from the model
p_values = mlr_2.pvalues

# List of insignificant features (p-value > 0.05)
insignificant_features = p_values[p_values > 0.05].index.tolist()

# Dropping these features from the dataset
X_3 = X_2.drop(columns=insignificant_features, errors='ignore')
```

```python
X_3
```

```python
X_train_3,X_test_3, y_train_3, y_test_3 = train_test_split(X_3,y,test_size=0.2,random_state=10)
```

```python
mlr_3 = sm.OLS(y_train_3, X_train_3)
mlr_3 = mlr_3.fit()
```

```python
mlr_3.summary()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_3, y_train_3)

# Make predictions
y_pred = rf_classifier.predict(X_test_3)

# Evaluate the model
accuracy = accuracy_score(y_test_3, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Display classification report
print(classification_report(y_test_3, y_pred))
```
