### Creating a new dataset for MLR

```python
terrMLR = terr[['extended', 'attacktype1_txt', 'targtype1_txt', 'gname', 'weaptype1_txt', 'INT_LOG', 'nkill']]
```

```python
terrMLR
```

```python
terr = terrMLR.dropna(how = "any")
```

```python
terr = terr[~terr.eq("Unknown").any(axis=1)]
```

```python
terr
```

```python
y = terr['nkill']
y
```

```python
X = terr.drop(['nkill'], axis = 1)
X
```

```python
X_1=pd.get_dummies(X,dtype=int)
X_1
```

```python
X_1=sm.add_constant(X_1)
X_1
```

```python
from sklearn.model_selection import train_test_split
X_train_1,X_test_1, y_train_1, y_test_1=train_test_split(X_1,y,test_size=0.2,random_state=10)
X_train_1.shape,X_test_1.shape, y_train_1.shape, y_test_1.shape
```

```python
mlr_1=sm.OLS(y_train_1,X_train_1)
```

```python
mlr_1=mlr_1.fit()
```
